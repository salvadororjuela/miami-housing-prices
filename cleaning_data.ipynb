```python
import pandas as pd
```

```python
# Data cleaning 
miami_housing_df = pd.read_csv("miami-housing.csv")
miami_housing_df.head()
```

```python
miami_housing_df.isnull().sum()
```

```python
miami_housing_df.dtypes
```

```python
# Convert the column names into a list to use later
column_names = miami_housing_df.columns.tolist()
column_names
```

```python
# Exclude "LATITUDE", "LONGITUDE", "PARCELNO", "avno60plus"
to_remove = ["LATITUDE", "LONGITUDE", "PARCELNO", "avno60plus"]

for att in to_remove:
    column_names.remove(att)
column_names
```

```python
miami_housing_df.head()
```

```python
# Find columns with values of 0 and replace them with the average value of the column. The 
# mean method won't be applied on the column "avno60plus" because it values are 1 = True, 0 = False
for column in column_names:
    # Get the mean value of the column
    mean_value = miami_housing_df[column].mean()
    # Replace values = 0 with the value of the mean in the column
    for attr in miami_housing_df:
        miami_housing_df[column].replace(0, int(mean_value), inplace = True)

miami_housing_df.head(10)
```

```python
miami_housing_df = miami_housing_df.head(1000)
```

```python
miami_housing_df.to_csv("cleaned_miami_housing.csv", index = False)
```
